```python
from fastai import *
from fastai.tabular import *
```

```python
features = features.assign(pts = df['pts'])
```

```python
features.columns
```

```python
procs = [Normalize]
dep_var = 'pts'
```

```python
cont_names = ['fga_100_decay',
'fg_pct_decay',
'fg2a_100_decay',
'fg2_pct_decay',
'fg3a_100_decay',
'fg3_pct_decay',
'fta_100_decay',
'ft_pct_decay',
'orb_100_decay',
'drb_100_decay',
'trb_100_decay',
'ast_100_decay',
'stl_100_decay',
'blk_100_decay',
'tov_100_decay',
'pf_100_decay',
'pace_decay',
'pm_100_decay',
'minutes_decay',
'age',
'days_rest',
'b2b',
'b2b_played',
'ln_rest_played',
'hca_dummy',
'team_change',
'opp_b2b',
'starter_last']
```

```python
data = (TabularList.from_df(features, cont_names=cont_names, procs=procs)
       .split_by_rand_pct()
       .label_from_df(cols=dep_var, label_cls=FloatList)
       .databunch())
```

```python
data.show_batch(rows=10)
```

```python
learn = tabular_learner(data, layers=[200,100], loss_func = FlattenedLoss(nn.L1Loss), metrics=mean_absolute_error)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(4, 1e-1)
```
